```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("1week.csv") 
```

```python
s=input("enter date in dd-mm-yyyy format")
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
m=data[(data['date']==s)]
```

```python
x=m['temp']
y=m['humid']
print(m)
```

```python
plt.plot(x,y,marker='o')
plt.title('data from the csvFile:Temperature and Humidity')
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.show()
```
